```python
#filtrar datos
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
class_names_split = ['Trouser','Pullover','Dress']
```

```python
i_t = np.where((train_labels == 1) | (train_labels == 2) | (train_labels == 3))
train_labels_split = train_labels[i_t]
train_images_split = train_images[i_t]

i_te = np.where((test_labels == 1) | (test_labels == 2) | (test_labels == 3))
test_labels_split = test_labels[i_te]
test_images_split = test_images[i_te]
```

```python
image_size = 28
num_labels = 3

def getReformatData(dataset, labels):
    dataset = dataset.reshape((-1, image_size * image_size)).astype(np.float32)
    data_normalized = dataset/255.0
    labels_df = pd.DataFrame(labels)
    labels_d = np.array(pd.get_dummies(labels_df[0]))
    return dataset, data_normalized, labels_d
```

```python
train_dataset, train_normalized, train_labels = getReformatData(train_images_split, train_labels_split)
test_dataset, test_normalized, test_labels = getReformatData(test_images_split, test_labels_split)
print('Training set', train_dataset.shape, train_normalized.shape, train_labels.shape)
print('Test set', test_dataset.shape, train_normalized.shape, test_labels.shape)
```

```python
def accuracy(predictions, labels):
    with g1.as_default() as g:
        acc, acc_op = tf.metrics.accuracy(labels=tf.argmax(labels, 1), 
                                      predictions=tf.argmax(predictions,1),name="accuracy")
        return acc_op
```

```python
#crear grafo
g1 = tf.Graph()

with g1.as_default() as g:
    tensor_x = tf.placeholder(tf.float32, shape=None, name="X")
    test_x = tf.placeholder(tf.float32, shape=None, name="Xtest")
    test_y = tf.placeholder(tf.float32, shape=None, name="Ytest")
    tensor_y = tf.placeholder(tf.float32,None,"Y")

    tf.set_random_seed(250)
    W = weights = tf.Variable(
    tf.truncated_normal([image_size * image_size, num_labels]))

    with tf.name_scope("logits") as scope:
        logits = tf.matmul(tensor_x, W,name="logits")

    with tf.name_scope("loss") as scope:
        loss = tf.reduce_mean(
            tf.nn.softmax_cross_entropy_with_logits(labels=tensor_y,logits= logits),name="loss")

    with tf.name_scope("gradient") as scope:
        gradiente = tf.gradients(loss,W,name="gradient")[0]

    with tf.name_scope("predictions") as scope:
        train_prediction = tf.nn.softmax(logits)
        test_prediction = tf.nn.softmax(tf.matmul(test_x, W))

    with tf.name_scope("accurracy") as scope:
        training_accurracy = accuracy(train_prediction,tensor_y)
        test_accurracy = accuracy(test_prediction,test_y)
```

```python
def training(X,Y,XT,YT,epochs,batch_size,lr,writer_name):
    loss_plt = []
    iteraciones = int(len(X)/batch_size)
    with g1.as_default() as g:
        
        with tf.name_scope("update") as scope:
            update_params = tf.assign(W,W - gradiente * lr,name="update_param")
        
        scalar = tf.summary.scalar('loss graph', loss)
        with tf.train.MonitoredSession() as sess:
            writer = tf.summary.FileWriter('./logs/'+datetime.datetime.now().strftime("%Y%m%d-%H")+str(writer_name), sess.graph)
            for e in range(epochs):
                for i in range(iteraciones):
                    offset = (e * batch_size) % (Y.shape[0] - batch_size)
                    batch_data = X[offset:(offset + batch_size), :]
                    batch_labels = Y[offset:(offset + batch_size), :]
                    feed_dict = {tensor_x : batch_data, tensor_y : batch_labels}
                    u, predictions = sess.run([update_params,train_prediction],feed_dict=feed_dict)
                feed_dict = {tensor_x: X, tensor_y: Y}
                l = sess.run(loss,feed_dict)
                loss_plt.append(l)
                summary = sess.run(scalar,feed_dict)
                writer.add_summary(summary, (e+1))
                if e % 100 == 0:
                    print("loss at step %d: %f" % (e, l))
                    print("Training accuracy: %.3f%%" % sess.run(training_accurracy,feed_dict))
            feed_dict = {test_x:XT,test_y:YT}
            print("Test accuracy: %.3f%%" % sess.run(test_accurracy,feed_dict))
            plt.plot(loss_plt)
            return sess.run(test_prediction,feed_dict),sess.run(W)
            
    
```

```python
def getPredictionIndex(image,W):
    test_prediction = tf.nn.softmax(tf.matmul(image, W))
    idx = tf.math.argmax(test_prediction[0])
    with tf.train.MonitoredSession() as sess:
        return(sess.run(idx))
```

```python
def graphImage(idx_image,model,data_set,normalize = False):
    image = data_set[idx_image]
    r_image = image.reshape((-1, 28 * 28)).astype(np.float32)
    pred = class_names_split[getPredictionIndex(r_image,model)] 
    return image,pred
```

```python
#seleccionando images al lazar, 5 
prueba = [2,10,100,330,400]
plt.figure(figsize=(20,20))
for i in range(5):
    img = graphImage(prueba[i],params_w,test_images_split)
    plt.subplot(1,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(img[0], cmap='gray')
    plt.xlabel('prediccin: '+img[1])
    plt.title('y real: '+class_names[test_labels_split[prueba[i]]])
plt.show()

    
```
